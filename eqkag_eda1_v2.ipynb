```python
plt.style.use('ggplot')
plt.figure(figsize=(18,10))
for i in range(1,7):
    file_sel = random.randint(0,(len(test_files)-1))
    df_tst = pd.read_csv(data_path + test_folder + '/' + test_files[file_sel])
    plt.subplot(2,3,i)
    plt.plot(df_tst['acoustic_data'])
    plt.xlabel('index')
    plt.ylabel('acoustic_data')
    plt.title('file:' + test_files[file_sel] + '  data_size=' + str(df_tst.shape))
```
